```
# part 1

groups = input_string.split('\n\n')

# print(groups)
yes_lengths = []

for group in groups:
  yes_sets = []
  for passenger in group.split():
    yesses = set()
    yesses.update(list(passenger))
    yes_sets.append(yesses)
  # Intersection is the magic word here. Makes a new set that is the overlap of 
  # the input sets. Expanding list to arguments with *.
  yes_lengths.append(len(set.intersection(*yes_sets)))

print(f"sum of counts: {sum(yes_lengths)}")
```
