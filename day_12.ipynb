```
# part 1
import numpy as np

instructions = [(x[0], int(x[1:])) for x in input_string.splitlines()]

# ship location
x = 0
y = 0
d = 0 # direction

for ins, val in instructions:
  # cardinal
  if ins == 'N': y += val
  if ins == 'S': y -= val
  if ins == 'E': x += val
  if ins == 'W': x -= val
  # forward
  if ins == 'F': 
    x += val*np.cos(np.deg2rad(d))
    y += val*np.sin(np.deg2rad(d))
  # turn
  if ins == 'R': d = d - val
  if ins == 'L': d = d + val
  d = d % 360
    
print(x, y, d)
print("Manhattan distance:", abs(x) + abs(y))
print("closest integer:", round(abs(x) + abs(y)))
```

```
# part 2
import numpy as np

instructions = [(x[0], int(x[1:])) for x in input_string.splitlines()]

# ship location
x = 0
y = 0
# waypoint location
wx = 10
wy = 1
d = 0 # direction

def rotate(point, about, d):
  θ = np.deg2rad(d)
  R = np.array([[np.cos(θ), -np.sin(θ)],
                [np.sin(θ),  np.cos(θ)]])
  
  return np.add(R.dot(
                  np.subtract(point, about)),
                about)

# print('ship:', x, y)
# print('waypoint:', wx, wy)
# print()

for ins, val in instructions:
  # cardinal
  if ins == 'N': wy += val
  if ins == 'S': wy -= val
  if ins == 'E': wx += val
  if ins == 'W': wx -= val
  # turn
  if ins == 'R': 
    wx, wy = rotate(point=(wx, wy), about=(0,0), d=-val)
  if ins == 'L': 
    wx, wy = rotate(point=(wx, wy), about=(0,0), d=val)
  # forward
  if ins == 'F':
    x, y = x + val*wx, y + val*wy

  # print('ship:', x, y)
  # print('waypoint:', wx, wy)
  # print()

print("Manhattan distance:", abs(x) + abs(y))
print("closest integer:", round(abs(x) + abs(y)))
```
