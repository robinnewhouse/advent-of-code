```
# part 2
import math
import numpy as np

def dec_to_bin(integer):
  '''convert integer to 36 bit binary'''
  binary = [0]*36
  quotient = integer
  for i in range(1, 36):
    binary[-i] = quotient%2
    quotient = math.floor(quotient/2)
  return(binary)

def bin_to_dec(binary):
  '''convert 36 bit binary to integer'''
  result = 0
  for i, x in enumerate(binary, start=1):
    result += x*2**(36-i)
  return result

def apply_mask(address, mask):
  masked_address = [0]*36
  for i, (a, m) in enumerate(zip(dec_to_bin(address), mask)):
    if m == '0': masked_address[i] = a
    if m == '1': masked_address[i] = 1
    if m == 'X': masked_address[i] = 'X'
  return masked_address

class Address():
  def __init__(self, address):
    self.address = address
    self.children = []
    self.fill_children()

  def fill_children(self):
    '''make a binary tree with possible children depending on binary value of X'''
    for i, x in enumerate(self.address):
      if x =='X':
        up = self.address.copy()
        up[i] = 1
        self.children.append(Address(up))
        down = self.address.copy()
        down[i] = 0
        self.children.append(Address(down))
        break

  def __str__(self):
    return ''.join([str(x) for x in self.address])

  def get_children(self):
    '''recursively get all children. returns nested in list'''
    if len(self.children) == 0:
      return self
    else:
      return [x.get_children() for x in self.children]

def calculate_addresses(masked_address):
  a = Address(masked_address)
  all_children = a.get_children()
  return [x.address for x in np.array(all_children).flatten()]

mask = ''
mem = {}
for line in input_string.splitlines():
  tokens = line.split(' = ')
  if tokens[0] == 'mask':
    mask = tokens[1]
  if tokens[0][0:3] == 'mem':
    address = int(tokens[0][4:-1])
    val = int(tokens[1])
    masked_address = apply_mask(address, mask)
    all_addresses = calculate_addresses(masked_address)
    for ad in all_addresses:
      mem[bin_to_dec(ad)] = val
  # print('n addresses', len(mem))
  
# print(mem)
print('solution', sum(mem.values()))
```
