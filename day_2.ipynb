```
# Part 1

valid = 0
invalid = 0

for line in input_string.splitlines():
  if not line: 
    continue
  tokens = line.split()
  min, max = [int(x) for x in tokens[0].split('-')]
  character = tokens[1][0]
  password = tokens[2]
  # print(min, max, character, password)

  count = password.count(character)
  if count >= min and count <= max:
    # password is valid
    valid +=1
  else:
    invalid +=1

print('Valid passwords', valid)
print('Invalid passwords', invalid)
```

```
# Part 2

valid = 0
invalid = 0

for line in input_string.splitlines():
  if not line: 
    continue
  tokens = line.split()
  pos1, pos2 = [int(x) for x in tokens[0].split('-')]
  character = tokens[1][0]
  password = tokens[2]

  # ^ is xor operator
  if (password[pos1-1] == character) ^ (password[pos2-1] == character):
    # password is valid
    valid +=1
  else:
    invalid +=1

print('Valid passwords', valid)
print('Invalid passwords', invalid)
```
