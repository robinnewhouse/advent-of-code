```
import numpy as np

def traverse(trees, right, down):
  x = 0
  encounter = 0
  # [start:end:step]
  for line in trees[::down]:    
    width = len(line)
    # repeats to the right
    if line[x%width] == '#':
      encounter += 1
    x += right
  return encounter

```

```
# part 1
# clean input. take off empty line at top.
clean_input = [line for line in input_string.split('\n') if line.strip() != '']

encounters = traverse(clean_input, right=3, down=1)
print(f"encountered {encounters} trees")
```

```
# part 2
# clean input. take off empty line at top.
clean_input = [line for line in input_string.split('\n') if line.strip() != '']

encounters = []
encounters.append(traverse(clean_input, right=1, down=1))
encounters.append(traverse(clean_input, right=3, down=1))
encounters.append(traverse(clean_input, right=5, down=1))
encounters.append(traverse(clean_input, right=7, down=1))
encounters.append(traverse(clean_input, right=1, down=2))
print(encounters)
print(np.product(encounters))
```
