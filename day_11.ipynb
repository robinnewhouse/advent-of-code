```
# part 1

from collections import Counter
import copy

input_grid = []

for line in input_string.splitlines():
  input_grid.append([seat for seat in line])

def count_filled_adjacent(grid, x, y):
  cnt = Counter()
  # check row up
  if y > 0:
    # three adjacent above
    cnt.update(grid[y-1][
                  max(x-1,0): # don't go past left edge
                  min(x+2,len(grid[y])) # dont go past right edge
                  ])
  if y < len(grid)-1:
  # three adjacent above
    cnt.update(grid[y+1][
                  max(x-1,0): # don't go past left edge
                  min(x+2,len(grid[y])) # dont go past right edge
                  ])
  if x > 0 :
    cnt.update(grid[y][x-1])
  if x < len(grid[y])-1:
    cnt.update(grid[y][x+1])
  return cnt['#']

def count_filled_lineofsight(grid, x, y):
  cnt = Counter()
  # check line of sight up
  yi = y - 1 
  while yi >= 0:
    if grid[yi][x] == '#':
      cnt.update('#')
      break
    if grid[yi][x] == 'L':
      break
    yi -= 1
  
  # check line of sight down
  yi = y + 1 
  while yi < len(grid):
    if grid[yi][x] == '#':
      cnt.update('#')
      break
    if grid[yi][x] == 'L':
      break
    yi += 1

  # check line of sight left
  xi = x - 1 
  while xi >= 0:
    if grid[y][xi] == '#':
      cnt.update('#')
      break
    if grid[y][xi] == 'L':
      break
    xi -= 1

  # check line of sight right
  xi = x + 1 
  while xi < len(grid[y]):
    if grid[y][xi] == '#':
      cnt.update('#')
      break
    if grid[y][xi] == 'L':
      break
    xi += 1

  # check line of sight up-left
  yi = y - 1 
  xi = x - 1 
  while xi >= 0 and yi >= 0:
    if grid[yi][xi] == '#':
      cnt.update('#')
      break
    if grid[yi][xi] == 'L':
      break
    yi -= 1
    xi -= 1

  # check line of sight up-right
  yi = y - 1 
  xi = x + 1 
  while xi < len(grid[y]) and yi >= 0:
    if grid[yi][xi] == '#':
      cnt.update('#')
      break
    if grid[yi][xi] == 'L':
      break
    yi -= 1
    xi += 1

  # check line of sight down-left
  yi = y + 1 
  xi = x - 1 
  while xi >= 0 and yi < len(grid):
    if grid[yi][xi] == '#':
      cnt.update('#')
      break
    if grid[yi][xi] == 'L':
      break
    yi += 1
    xi -= 1

  # check line of sight down-right
  yi = y + 1 
  xi = x + 1 
  while xi < len(grid[y]) and yi < len(grid):
    if grid[yi][xi] == '#':
      cnt.update('#')
      break
    if grid[yi][xi] == 'L':
      break
    yi += 1
    xi += 1

  return cnt['#']


def print_grid(grid):
  for y in range(len(grid)):
    for x in range(len(grid[y])):
      print(grid[y][x], end='')
    print()
  print()

for i in range(100):
  changed = False
  current_grid = copy.deepcopy(input_grid)
  for y in range(len(current_grid)):
    for x in range(len(current_grid[y])):
      if count_filled_lineofsight(input_grid,x,y) == 0:
        if current_grid[y][x] == 'L':
          current_grid[y][x] = '#'
          changed = True
      if count_filled_lineofsight(input_grid,x,y) >= 5:
        if current_grid[y][x] == '#':
          current_grid[y][x] = 'L'
          changed = True
  input_grid = current_grid
  # print(i)
  if changed == False:
    break

print(i)
print_grid(input_grid)
cnt = Counter([seat for row in input_grid for seat in row])
print(cnt['#'], 'occupied seats')
```
